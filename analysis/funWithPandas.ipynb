I was reading about Pandas over the weekend, and came across some nice features that will help with your analysis, what Pandas calls "hierarchical indexing", "reindexing", "grouping", and "unstacking".

Here's a demo.

```python
%pylab inline
import pandas as pd
```

```python
# set the index.  Since I'm giving it a list of columns, the result is a heirarchical index...
kr = k.set_index(['day','osc','chan','shot'])
```

```python
# heirarchical indexing, for example, calculating means with .mean() to simplify the output.
a = kr.amp

print a.mean() # the mean of all hits on all days
print a.ix[22].mean() # day 22, mean of all hits.
print a.ix[22].ix[2].mean() # day 22, mean of hits in scope 2
print a.ix[22].ix[2].ix[3].mean() # day 22, scope 2, channel 3
print a.ix[22].ix[2].ix[3].ix[9:15].mean() # day 22, scope 2, channel 3, shots 9 through 15.

# note that none of these indexing results are unique.  Even the last one has multiple hits.
# This is useful for slicing your data into specific shots or ranges of shots
```

```python
# groupby splits your data up into groups.  level=0 here says to use the index for the groups.  Otherwise, it wants the name of a column.
# the result of groupby(...) has a .min() function that just takes the minimum value of the entries of the group.
a22 = a.ix[22].ix[2].ix[2].groupby(level=0).min()
a23 = a.ix[22].ix[2].ix[3].groupby(level=0).min()
```

```python
# reindexing makes the index of one bit of data conform to the index of another,
# adding NaNs whenever there are missing values, dropping values when not needed.
# here, missing data implies the detector was not hit, so 0 is better than NaN,
# so I use that for fill_value.
a22r = a22.reindex(a23.index,fill_value=0)
a23r = a23.reindex(a22.index,fill_value=0)
```

```python
# what's nice here is you can make some plots...
scatter(-a22,-a23r)
title("day 22 biggest pulses compared by channel")
xlabel("scope 2 channel 2 biggest pulse")
ylabel("scope 2 channel 3 biggest pulse")
```

```python
scatter(-a22r,-a23)
title("day 22 biggest pulses compared by channel")
xlabel("scope 2 channel 2 biggest pulse")
ylabel("scope 2 channel 3 biggest pulse")
```

Note that these plots have a lot of points in common.  The only differences are the ones that are dropped by the reindexing process, which you can see along the x=0 and y=0 axis.

You can do better... "unstack".

```python
# take day 22, split it up into different scopes and shots, take the biggest pulse in each group...
x = a.ix[22].groupby(level=['osc','shot']).min()
# then, unstack.
xu = x.unstack(0)
```

```python
print x.head(5)
print 
print xu.head(5)
```

"unstack(0)" converted the values of the first level of the index into columns, in this case one columns for each scope.  The remaining parts of the index, i.e. the shot number, stays part of the index.  Since not all shots have hits, unstack adds NaNs.  As I said before, 0 is more useful than NaN....

```python
# replace NaNs with 0.
xu = xu.fillna(0)
```

```python
# now, plot.
scatter(-xu[2],-xu[3])
xlabel("scope 2 biggest pulse")
ylabel("scope 3 biggest pulse")
title("day 22 biggest pulses compared by scope")
```
